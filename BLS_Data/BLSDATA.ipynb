```python
## Import dependencies

import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress
import csv
from api_keys import BLS_API_KEY

## Define my output file
output_data_file = "BLS_Data/unemployment_monthly_2018_to_2020.csv"
```

```python
## Calling BLS API
url = "https://api.bls.gov/publicAPI/v2/timeseries/data/"
apiKEY = BLS_API_KEY
seriesName = "LASST420000000000003"
params = {"seriesid" : seriesName, "startyear" : 2018, "endyear" : 2020, "registrationkey" : apiKEY}
BLSData = requests.post(url, data=params)
BLSData
```

```python
## Setting the output from my loop later to turn into a data frame to write to a .csv
year = []
month = []
monthNumber = []
unemploymentRate = []
```

```python
## Converting my API data into a JSON
BLSJSON = BLSData.json()
## Getting just the data I need
BLSJSONDATALIST = BLSJSON["Results"]["series"][0]["data"]

## 12 months per year over 3 years is 36 months. The data is monthly data so we run the loop 36 times to get all the data
for data in range(0, 36):
    year.append(BLSJSONDATALIST[data]["year"])
    monthNumber.append(BLSJSONDATALIST[data]["period"])
    month.append(BLSJSONDATALIST[data]["periodName"])
    unemploymentRate.append(BLSJSONDATALIST[data]["value"])
```

```python
## Finally creating my dataframe for unemployment rate
BLS_DF = pd.DataFrame({"Year" : year,
                      "Month" : month,
                      "Month #" : monthNumber,
                      "Unemployment Rate" : unemploymentRate})
BLS_DF
```

```python
## Finally I export as a .csv
BLS_DF.to_csv(output_data_file, index = False)
```
