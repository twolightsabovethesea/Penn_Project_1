```python
import pandas as pd
```

```python
unemployment2019_df =pd.read_csv('unemployment_2019.csv')
```

```python
unemployment2019_df.head()
```

```python
clean_unmployment2019_df = unemployment2019_df[['Areaname','Areatype','Periodyear','Timeperiod','Laborforce','Emplab','Unemp','Unemprate']]
clean_unmployment2019_df.head()
```

```python
unemployment2020_df =pd.read_csv('unemployment_2020.csv')
```

```python
unemployment2020_df.head()
```

```python
clean_unmployment2020_df = unemployment2020_df[['Areaname','Areatype','Periodyear','Timeperiod','Laborforce','Emplab','Unemp','Unemprate']]
clean_unmployment2020_df.head()
```

```python
cleanunemployment20192020_df = pd.concat([clean_unmployment2019_df, clean_unmployment2020_df], axis=0)
```

```python
cleanunemployment20192020_df.head()
```

```python
cleanunemployment20192020_df.tail()
```

```python
countyunemployment_df = cleanunemployment20192020_df.loc[cleanunemployment20192020_df['Areatype'] == 4]
```

```python
countyunemployment_df.head()
```

```python
countyunemployment_df["Areaname"] = countyunemployment_df["Areaname"].str.split(" ").str[0]
countyunemployment_df.head()
```

```python
countyunemployment_df
```

```python
del countyunemployment_df['Areatype']
```

```python
countyunemployment_df.head()
```

```python
countyunemployment_df.rename(columns={'Areaname':'County Name','Periodyear':'Year','Timeperiod':'Months'})
```

```python
countyunemployment_df.to_csv('countyunemployment.csv', index=False, header = True)
```
