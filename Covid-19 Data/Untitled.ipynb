```python
## Import dependencies
import pandas as pd
import scipy.stats as stats
import numpy as np

## Importing data
filepath = "COVID-19_Aggregate_Cases_Current_Daily_County_Health.csv"
outputPath = "Covid-19_Monthly_Data"
```

```python
## Reading in our file to a dataframe
covidDF = pd.read_csv(filepath)
```

```python
## Sorting by county and date
covidCounty = covidDF.sort_values(["Jurisdiction", "Date"])
covidCounty.head()
```

```python
## Converting date to datatime type in order to convert to monthly data
covidCounty["Date"] = pd.to_datetime(covidCounty["Date"])
covidCounty.head()
```

```python
covidCounty["Month"] = covidCounty["Date"].dt.to_period("M")
```

```python
covidCounty.head()
```

```python
groupDF = covidCounty.groupby(["Jurisdiction", "Month"])
```

```python
groupedSUM = groupDF.sum()
groupedSUM.reset_index(inplace = True)
regularGroup = groupDF.head(1000000000)
groupedSUM
```

```python
summedDF = pd.DataFrame({"County" : groupedSUM["Jurisdiction"],
                        "Date" : groupedSUM["Month"],
                        "Monthly Cases" : groupedSUM["New Cases"],
                        "Longitude" : regularGroup["Longitude"],
                        "Latitude" : regularGroup["Latitude"]}).dropna()
summedDF
```

```python
summedDF.to_csv(outputPath)
```
