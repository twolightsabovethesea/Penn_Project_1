```python
#import dependencies
import pandas as pd
```

```python
# read in CSV data
statewide_2018 = pd.read_csv("CRASH_2018.csv")
statewide_2018.head()
```

```python
#read in more CSV data
flag_2018 = pd.read_csv("FLAG_2018.csv")
flag_2018.head()
```

```python
# read in county name csv

counties_df = pd.read_csv("../Crash_Data_2019/county.csv")
counties_df
```

```python
#remove unneeded columns from statewide data

statewide_cleaned = statewide_2018[["CRN", "COUNTY", "CRASH_MONTH", "DEC_LAT", "DEC_LONG", "FATAL_COUNT"]]
statewide_cleaned.head()
```

```python
#clean flag data frame
flag_cleaned = flag_2018[["CRN", "ALCOHOL_RELATED", "DRINKING_DRIVER", "UNDERAGE_DRNK_DRV"]]
flag_cleaned.head()
```

```python
#merge the statewide and flag dataframes
merged_df = pd.merge(statewide_cleaned, flag_cleaned, on="CRN",how="outer")
```

```python
merged_df.head()
```

```python
#check the number of rows before dropping unneeded rows
merged_df["CRN"].count()
```

```python
#remove rows with data on crashes not related to alcohol

zero = [0]
alcohol_related = merged_df[merged_df["ALCOHOL_RELATED"].isin(zero)==False]
```

```python
# confirm that unneeded rows were dropped
alcohol_related["CRN"].count()
```

```python
#rename column before merge

counties_df.rename(columns={"COUNTY ID": "COUNTY"}, inplace=True)
counties_df.head()
```

```python
#merge alcohol_related_df with counties_df to add County Name

dui_2018 = alcohol_related.merge(counties_df, how="left", on="COUNTY")
dui_2018
```

```python
dui_2018.to_csv("dui_2018_csv",index=False,header=True)
```
